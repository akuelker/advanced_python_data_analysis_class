```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('LemonadeStandWeekend.csv')
df = df.fillna(0)

df.columns = ['Day', 'Time', 'Kid', 
                'Sales', 'Donations', 'Lemonade', 'Cookies']
df.plot(kind = 'scatter', x = 'Time', y = 'Sales', color = 'red', title = 'Sales')
plt.savefig('LemonadeScatter.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('LemonadeStandWeekend.csv')
df = df.fillna(0)
df.columns = ['Day', 'Time', 'Kid', 
                'Sales', 'Donations', 'Lemonade', 'Cookies']
df.plot(kind = 'line', x = ['Day','Time'], y = ['Lemonade', 'Cookies'], title = 'Food Sold')
plt.savefig('LemonadeLine.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('LemonadeStandWeekend.csv')
df = df.fillna(0)
df.columns = ['Day', 'Time', 'Kid', 
                'Sales', 'Donations', 'Lemonade', 'Cookies']

df.groupby(['Day']).plot(kind = 'barh', x = ['Day','Time'], y = ['Sales', 'Donations'],  title = 'Day to Day Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv('LemonadeStandWeekend.csv')
df = df.fillna(0)
df.columns = ['Day', 'Time', 'Kid', 
                'Sales', 'Donations', 'Lemonade', 'Cookies']

sns.barplot(df.Time, df.Sales)
plt.savefig('LemonadeEachDay.jpg')
plt.show()
```

```python
fig, ax = plt.subplots()
percentages = [sum(df.Sales), sum(df.Donations)]
labels = ['Sales', 'Donations']
ax.pie(percentages, labels = labels)
ax.axis('equal')
plt.savefig('PieMoney.jpg')
plt.show()
```
