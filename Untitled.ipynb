```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('PopUp Toy Store4_5.csv')
data = data.fillna(0)
data.columns = ['Week', 'Day', 'Airplanes', 
                'Sticks', 'Weather']
data.plot (kind = 'bar', x = 'Day', y = 'Airplanes', 
         color = 'purple', title = 'Airplanes Sold')
plt.savefig('Airplanes1.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.hist(data.Airplanes, label = 'Airplanes',color = 'blue', alpha = .3)
fig = sns.rugplot(data.Airplanes,  color = 'blue')
fig.figure.suptitle ('Airplanes histogram and rugplot', fontsize = 20)
plt.grid(which='major', alpha=0.5, color='red', linestyle='-', linewidth=2)
plt.savefig("Rugplot.jpg")
plt.show()
```
