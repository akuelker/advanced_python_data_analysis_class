```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('PopUp Toy Store4_5.csv')
data = data.fillna(0)
data.columns = ['Week', 'Day', 'Airplanes', 
                'Sticks', 'Weather']
data.plot (kind = 'bar', x = 'Day', y = 'Airplanes', 
         color = 'purple', title = 'Airplanes Sold')
plt.savefig('Airplanes1.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('PopUp Toy Store4_5.csv')
data = data.fillna(0)
data.columns = ['Week', 'Day', 'Airplanes', 
                'Sticks', 'Weather']
fig, ax = plt.subplots()
groups = data.groupby(['Week']).plot(x = 'Day', y = 'Airplanes', ax = ax, legend = False)
plt.savefig('Airplanes2.jpg')
plt.show()
```

```python
fig, ax = plt.subplots()
percentages = [sum(data.Airplanes), sum(data.Sticks)]
labels = ['Airplanes', 'Sticks']
ax.pie(percentages, labels = labels)
ax.axis('equal')
plt.savefig('ToyPie.jpg')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = pd.read_csv('PopUp Toy Store4_5.csv')
data = data.fillna(0)
data.columns = ['Week', 'Day', 'Airplanes', 
                'Sticks', 'Weather']
sns.barplot(data.Day, data.Airplanes)
plt.savefig('BarAirplanes.jpg')
plt.show()
```

```python
sns.barplot(data.Day, data.Sticks)
plt.savefig('BarSticks.jpg')
plt.show()
```

```python
sns.boxplot(data.Day, data.Airplanes)
plt.savefig('BoxPlot.jpg')
```

```python
sns.kdeplot(data.Week, data.Airplanes)
plt.savefig('kdeAirplanesWeek.jpg')
```

```python
sns.kdeplot(data.Airplanes)
plt.savefig('kdeAirplanes.jpg')
```

```python
fig = sns.kdeplot(data.Airplanes, shade = True, 
                  color = 'blue')
fig = sns.kdeplot(data.Sticks, shade = True, 
                  color = 'red')
fig.figure.suptitle("Toys", fontsize = 20)
plt.xlabel('Toys', fontsize=18)
plt.savefig('kdeAirplanesAndSticks')
```

```python
fig = sns.distplot(data.Sticks, 
                   label = 'Sticks', color = 'red')
fig = sns.distplot(data.Airplanes, 
                   label = 'Airplanes', color = 'blue')
fig.figure.suptitle("Distribution of Toys", fontsize = 20)
plt.xlabel('Toys', fontsize=18)
plt.legend()
plt.savefig('distAirplanesAndSticks')
```

```python
fig = plt.hist(data.Airplanes, label = 'Sticks',color = 'blue', alpha = .3)
fig = sns.rugplot(data.Airplanes,  color = 'blue')
fig.figure.suptitle ('Airplanes histogram and rugplot', fontsize = 20)
plt.grid(which='major', alpha=0.5, color='red', linestyle='-', linewidth=2)
plt.savefig("Rugplot.jpg")
plt.show()
```
