```python
import pandas as pd
acmeColNames = ['Invoice', 'District', 'Sale','Date']
acmeIndex = ['Invoice']
salesAcme = pd.read_csv('AcmeWidgetSales.csv',
                        header = 0,
                        index_col = acmeIndex,
                       names = acmeColNames)
salesAcme
```

```python
import pandas as pd
salesAcme2 = salesAcme.set_index(['Date', 'District'])
salesByDate = salesAcme2['Sale'].sum(level =['Date'])
```

```python
import matplotlib.pyplot as plt
salesByDate.plot(kind = 'bar', x = 'Date', y = 'Sale',  title = 'Widgets')
plt.show()
```

```python
#Compare districts to each other

myFigure, myAxis = plt.subplots()
#groups = salesAcme.groupby(
 #   ['District']).plot(x = 'Date', y = 'Sale', ax = myAxis)
groups = salesAcme.groupby(
    ['District']).plot(x = 'Date', y = 'Sale', ax = myAxis)
myAxis.legend(labels = groups.keys(), title = 'Districts')
plt.show()
```

```python

salesByDistrictDate = salesAcme2.groupby(['District', 'Date'], as_index=True)['Sale'].sum()
#Compare districts to each other

myFigure, myAxis = plt.subplots()

groups = salesByDistrictDate.groupby(
    ['District']).plot(x = 'Date', y = 'Sale', ax = myAxis)
myAxis.legend(labels = groups.keys(), title = 'Districts')
plt.xticks(rotation = 90)
#Get the dates
dateGroups = salesByDistrictDate.groupby(['Date'])
plt.xlabel = dateGroups
plt.show()
```
